# iterations in python
1. iterations are used to work on repetitive tasks or repetitive performances
2. python having  2 types of iterations 
3. `for` loop
4. `while` loop


```python - for loop
for iter_variable in sequence
    print('for is keyword to start the iterations')
    print('iter_variable is to get values from sequence')
    print('in is keyword is valid sequence present orn not')
    print('sequence is collection like range,string,list....etc')
```

```python - while loop
while(condition):
    print ('while is keyword to valid condition based on data or values')

````python
# write a program to print 1 to 20 numbers using for loop
for i in range(1,21):
    print(i)
````

````python
# write a program to iterate string using for loop
info="welcome to python program"
for char in info:
    print(char)
````

````python
#write a program to print even numbers in between 50 to 100 using for lopp
for n in range(50,101):
    if (n%2 == 0):
        print(n,end=",")
````

````python
for n in range(50,101):
    if (n%2 == 1):
        print(n,end=",")
````

````python
#write  a program to find the vowels form string using for loop
data="python programming"
for char in data:
    if(char  not in "aeiouAEIOU"):
        print(char)
````
