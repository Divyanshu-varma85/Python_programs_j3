```python
# write a program to find the given number is odd number?
number=int(input("enter number here!"))
if(number%2==0):
    print(number,"is even number")
else:
    print(number,"is odd number")
```

```python
# write a program to find if the number is negative or positive or 0
number=int(input("enter number here!"))
if(number>0):
         print(number,"is positive number")
elif(number<0):
           print(number,"is negative number")
else:
           print(number,"is zero")
```

```python
"""
write a program to find the grades based on student marks
1.if marks in between 90 to 100:A grade
2.if marks in between 90 to 80: B grade
3.if marks in between 80 to 70: C grade
4.if marks in between 70 to 60: D grade
5. if marks in between 60 to 50:E grade 
6.if marks in between 50 to 40:F grade
7.if other than above marks output is Please enter correct marks

"""
marks=int(input("enter student marks here"))
if(marks>=90 and marks ==100):
    print(marks,"A grade")
elif(marks>=90 and marks ==80):
    print(marks,"B grade")
elif(marks>=80 and marks == 70):
    print(marks,"C grade")
elif(marks>=70 and marks==60):
    print(marks,"D grade")
elif(marks>=60 and marks==50):
    print(marks,"is E grade")
elif(marks>=50 and marks==40):
    print(marks,"is F grade")
else:
    print("please enter correct marks")
```
